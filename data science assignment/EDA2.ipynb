```python
import pandas as pd

# Load the dataset
file_path = r"D:\Excelr\Data Science\Data Science Assignment\EDA2\EDA2\adult_with_headers.csv"
df = pd.read_csv(file_path)
data_info = df.info()
summary_stats = df.describe(include='all')
missing_values = df.isnull().sum()

data_info, summary_stats, missing_values
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Selecting numerical columns
numerical_cols = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
# Applying Standard Scaling
standard_scaler = StandardScaler()
standard_scaled_df = pd.DataFrame(standard_scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)
# Applying Min-Max Scaling
min_max_scaler = MinMaxScaler()
min_max_scaled_df = pd.DataFrame(min_max_scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)
# Displaying the scaled datasets
standard_scaled_df.head(), min_max_scaled_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
# Identifying categorical columns with less than and more than 5 categories
categorical_cols = df.select_dtypes(include=['object']).columns
one_hot_cols = [col for col in categorical_cols if df[col].nunique() < 5]
label_encode_cols = [col for col in categorical_cols if df[col].nunique() >= 5]
# Applying One-Hot Encoding to categorical columns with less than 5 categories
one_hot_encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' to avoid multicollinearity
one_hot_encoded_df = pd.DataFrame(one_hot_encoder.fit_transform(df[one_hot_cols]), 
                                  columns=one_hot_encoder.get_feature_names_out(one_hot_cols))
# Applying Label Encoding to categorical columns with more than 5 categories
label_encoder = LabelEncoder()
# Applying Label Encoding to categorical columns with more than 5 categories
label_encoded_df = df[label_encode_cols].apply(label_encoder.fit_transform)
# Concatenating the original dataframe (dropping the encoded categorical columns) with the new encoded columns
encoded_df = pd.concat([df.drop(columns=one_hot_cols + label_encode_cols), one_hot_encoded_df, label_encoded_df], axis=1)

# Display the encoded dataframe
encoded_df.head()
```

```python
# Creating the 'work_hours_category' feature
def categorize_hours(hours):
    if hours < 30:
        return 'part-time'
    elif 30 <= hours <= 40:
        return 'full-time'
    else:
        return 'overtime'

df['work_hours_category'] = df['hours_per_week'].apply(categorize_hours)

# Creating the 'capital_gain_loss_ratio' feature (avoiding division by zero)
df['capital_gain_loss_ratio'] = df.apply(lambda row: row['capital_gain'] / (row['capital_loss'] + 1), axis=1)

# Displaying the new features
df[['hours_per_week', 'work_hours_category', 'capital_gain', 'capital_loss', 'capital_gain_loss_ratio']].head()
```

```python
import numpy as np

# Applying log transformation to the skewed 'capital_gain' feature (adding 1 to avoid log(0))
df['log_capital_gain'] = np.log(df['capital_gain'] + 1)

# Displaying the original and transformed 'capital_gain' columns
df[['capital_gain', 'log_capital_gain']].head()
```

```python
from sklearn.ensemble import IsolationForest

# Select numerical columns for outlier detection
numerical_data = df.select_dtypes(include=['int64', 'float64'])
# Initialize and fit the Isolation Forest model
iso_forest = IsolationForest(contamination=0.05, random_state=42)
outliers = iso_forest.fit_predict(numerical_data)
# Label the outliers (outliers are labeled as -1, inliers as 1)
df['outlier'] = outliers
# Removing the outliers from the dataset
df_cleaned = df[df['outlier'] == 1].drop(columns=['outlier'])
# Display the result after removing outliers
df_cleaned.head(), df.shape, df_cleaned.shape
```

```python
import ppscore as pps

# PPS matrix for the dataset
pps_matrix = pps.matrix(df_cleaned)

# Display the matrix as a pivot table for easier interpretation
pps_matrix[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
```
